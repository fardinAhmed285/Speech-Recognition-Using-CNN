The problem domain that I am going to address is called Speech recognition.

Speech recognition by machines is being researched for more than half a century. One of the challenges of voice recognition is to reliably understand words in distorting conditions. Some of its applications are voice assistance, home automation, vehicle navigation, gaming etc. Nowadays, with the accelerated growth of big data and higher computing power, applications that were categorized as unreal in the past are becoming a reality. The process that generates speech from thoughts can be described in five stages. At first the thought goes into the word sequence generation stage. Then comes the speech generation and speech production stage. The fourth stage is called the Acoustic channel where the speech gets differentiated from the noise. At last, it produces the final speech. By separating speech and interfering signals with the help of spectral and temporal separation, the problem posed by noisy background can be mitigated. That’s where AI is going to help us. Until recently, ASR acoustic models were developed by using Gaussian mixture model hidden Markov models (GMM-HMMs). However, a new modelling technique which uses deep learning has been developed. Its called context-dependent deep neural network hidden Markov model (CD-DNN-HMM). In the late 1980s, after suffering a huge blow by the paper written by Minsky and Papert about the inability of representing the simple XOR functions, neural nets were being used for ASR. Some of the labs were using hybrid HMM-ANN techniques. In these techniques HMM was used to look for the word strings from the ANN and the ANN was used as a phonetic probability estimator. Acoustic-phonetic approach, Generative learning approach --HMM-GMM, Deep-learning based approach are some of the conventional ASR methods. Below is a brief description of these methods,

**Acoustic-phonetic approach**: The acoustic-phonetic method starts by studying the speech sounds without considering their meaning. Then, it uses strong evidence to turn this information into specific elements that describe how speech sounds are made. Next, it divides the speech into stable sound areas and labels each one with a phonetic symbol to show the different speech sounds. Finally, it tries to figure out meaningful words by using these labelled sound groups created earlier.

**Generative learning approach**: GMM-based HMMs are like building blocks of the generative learning approach. They use mixtures of Gaussians to understand the order of sounds in speech. In old-style speech systems, these models help figure out the patterns in how words follow each other.

**Deep-learning based approach**: Speech recognition systems often use neural networks like CNNs, RNNs, LSTMs, and transformer-based methods to create complete systems from start to finish.

AI is used in speech recognition because there are many problems that needs to be solved to build a system that can recognize speech. The vocabulary size, different kind of speech (Isolated, Discontinuous, and Continuous), adverse conditions are some of them.

**Vocabulary size**: It's easy to tell words apart when there are only a few options. We use small speech recognition systems for things like voice commands, dialing by voice, and giving simple instructions. But for understanding longer sentences and transcribing live speech in videos, we need bigger speech recognition systems that can recognize thousands of different words.

**Isolated, Discontinuous, and Continuous Speech**: At first, recognizing single words is done with voice data that's separate. When words are deliberately split by silence in sentences, it's called discontinuous speech. Continuous speech means sentences are spoken naturally. It's easier to recognize words in isolated or discontinuous speech because they're well said and separated by clear pauses. But in continuous speech, it gets harder because the breaks between words are unclear, and how words are said can change. When the system deals with casual speech, like coughs, emotions, or starting sentences and not finishing them, it becomes even more challenging.

**Adverse conditions**: Typically, a system gets trained using clear sound data. But there are situations that can really mess up how well the system works. These include things like background noise, differences in how the sound is picked up, and any extra noise that comes through the connection. (Kumar and Renuka, 2023)

## CNN

Within the CNN architecture, there exist input and output layers, along with intermediary opaque layers. These opaque layers, often referred to as hidden layers, play a crucial role in identifying distinct features within the data. Components like the Linear Unit for Convolution, Pooling, and Rectifier enhance the network's learning speed and efficiency. Once these layers have effectively identified features, the CNN architecture moves forward into the classification stage, where it assigns labels or categories to the detected features. The convolutional layers in a CNN consist of diverse filters, often referred to as kernels, that can be adjusted during training. These filters move systematically across the entire input, generating an output function map for each filter. By applying the same filter to various positions across the input, it effectively scans through the input space. In convolutional layers, smaller filter sizes are typically employed, enabling the detection of local features that hold significance in data like images and sound. The function map produced by the previous layer comprises stacked convolutional layers, combining more complex features from the input data. Convolutional layers focus on capturing local features, whereas pooling layers merge related features by selecting the maximum value (max pooling) or averaging values (average pooling). This step reduces sensitivity to minor variations and distortions in the features by shrinking the feature map size. Positioned after one or more convolutional layers, pooling layers continually reduce the input size as the network grows deeper. Each reduction in input size, coupled with the pooling operation, also diminishes the computations needed within the network.
(Abdulazeez and Taher, 2021)

## RNN


Traditional speech recognition systems involve several algorithms and processing stages. However, deep learning addresses this complexity by employing a language model, which, despite its simplicity, yields superior outcomes. This approach proves particularly effective in handling issues related to noise and variations in speakers. Yet, reaping the benefits of deep learning poses certain challenges. One such challenge involves devising strategies to construct extensive, well-labeled datasets. Additionally, there's a need to train networks capable of effectively leveraging these datasets for optimal performance. We also need to make sure that the labeled data is aligned perfectly with the text transcripts of the speech. RNN is trained to get speech spectrogram as input and generate English text. We must pair spoken words (x) with their corresponding labels (y) in the training dataset. Each spoken word x is a sequence of audio features collected over time, where each 'slice' of time will capture specific sound details. These features are represented using spectrograms, where x represents the power of various frequency ranges at each time instance. The aim of the Recurrent Neural Network (RNN) is to process these sequences of audio features and transform them into a sequence of probabilities associated with the characters forming the transcribed word. Essentially, we're tasking the RNN with predicting the likelihood of each letter or symbol (such as spaces or apostrophes) being the next part of the spoken word. RNNs can be employed to capture temporal dependencies in the learned features. That will be a bi-directional recurrent layer. It will contain a set with forward recurrence and a set with backward recurrence. The last layer of the network is going to use information from both earlier layers, combining them to make predictions. This layer takes inputs from forward and backward units and does some calculations using weights and biases. The output layer will utilize a standard softmax function to predict the probabilities of different characters for each moment in time and for each character in the alphabet. This prediction will be based on the combined information from the previous layer's calculations. The mathematical operations involve exponential functions and a weight matrix to make these predictions for each character in the alphabet at every time step. We can use dropout at a rate of 5% in the feed forward layers to reduce further variance. (Hannun et al., 2014)

## HMM


The Hidden Markov Model (HMM) is a powerful statistical framework that captures the essence of systems evolving through a sequence of states, each emitting distinct observations. In the realm of speech recognition, these states take on the role of representing phonemes or elemental units of sound, connected by a web of probabilities governing their transitions. These transitions contain the dynamic movement from one state to another, shaping the course of the system's evolution.
Each state is associated with invaluable observations, typically known as acoustic features such as Mel-frequency cepstral coefficients (MFCCs). These observations provide a view into the unique acoustic characteristics defining each state, offering insights into the sound associated with a particular phoneme or sound unit.
The heart of the HMM lies in its comprehensive components, from the probabilities guiding state transitions and emissions to the crucial definition of initial state probabilities and observations. To refine the model's parameters and optimize its accuracy, the Baum-Welch Algorithm, employing the principles of Expectation-Maximization, iteratively adjusts these probabilities based on observed data until there are no changes, ensuring a closer alignment with the reality of the spoken language.
Decoding the sequences of observed data falls into the domain of the Viterbi Algorithm, a powerful tool that computes the most probable sequence of states, unlocking the underlying structure of the observed data and revealing the path that maximizes the overall probability.
The diverse applications of HMMs span across speech and speaker recognition, as well as language identification. However, challenges arise in managing the complexity of the model and accommodating the variability found in speech due to diverse accents, environmental conditions, and individual speaking styles.
Yet, in the ever-evolving landscape of speech recognition, the legacy of HMMs remains foundational. While they set the groundwork for this field, modern systems pivot toward neural network-based methodologies, particularly deep learning. Deep neural networks (DNNs) and recurrent neural networks (RNNs) surpasses HMM-based systems by capturing patterns and relationships within sequential speech data. Their ability to comprehend complex speech dynamics yields more robust and accurate speech recognition systems, charting new frontiers in this ever-evolving domain. (Rabiner and Juang, 2005)



The hidden Markov model (HMM) is a statistical tool used to understand how different sounds in a language can vary in how they sound. It uses a special way of describing these variations, kind of like fitting puzzle pieces together to match how people talk. By making small improvements repeatedly, it tries to perfectly match how these sounds change in actual speech until it gets good at it. (Jelinek, 2022)

```python
labels=['two', 'wow', 'up', 'tree', 'zero', 'six', 'three', 'stop', 'yes', 'left', 'on', 'nine', 'marvin', 'off', 'no', 'one', 'sheila', 'seven', 'right', 'dog', 'bird', 'down', 'go', 'house', 'four', 'happy', 'cat', 'five', 'eight', 'bed']
```

```python
#Duration of recordings
duration_of_recordings=[]
for label in labels:
    waves=[f for f in os.listdir(train_audio_path+'/'+label) if f.endswith('.wav')]
    for wav in waves:
        sample_rate,samples = wavfile.read(train_audio_path+'/'+label+'/'+wav)
        duration_of_recordings.append(float(len(samples)/sample_rate))

plt.hist(np.array(duration_of_recordings))
```

```python
#Data preproccesing- resampling and removing shorter commands

all_wave=[]
all_label=[]
for label in labels:
    print(label)
    waves=[f for f in os.listdir(train_audio_path+'/'+label) if f.endswith('.wav')]
    for wav in waves:
        samples,sample_rate = librosa.load(train_audio_path+'/'+label+'/'+wav, sr=16000)
        target_sample_rate = 8000
        num_samples = int(len(samples) * target_sample_rate / sample_rate)
        samples = resample(samples, num_samples)
        if(len(samples)==8000):
            all_wave.append(samples)
            all_label.append(label)
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(all_label)
classes = list(le.classes_)
```

```python
#One hot vector
from keras.utils import to_categorical
y=to_categorical(y,num_classes=len(labels))
```

```python
all_wave=np.array(all_wave).reshape(-1,8000,1)
```

```python
from sklearn.model_selection import train_test_split

# Split data into 80% train and 20% combined validation/test
x_tr, x_combined, y_tr, y_combined = train_test_split(np.array(all_wave), np.array(y), stratify=y, test_size=0.2, random_state=777, shuffle=True)

# Split the combined 20% into 50% validation and 50% test
x_val, x_test, y_val, y_test = train_test_split(x_combined, y_combined, test_size=0.5, random_state=777, shuffle=True)
```

```python
#Build a model using Keras
from keras.layers import Dense, Dropout, Flatten, Conv1D, Input, MaxPooling1D
from keras.models import Model
from keras import backend as K
K.clear_session()

inputs = Input(shape=(8000, 1))

# First Conv1D Layer
conv1 = Conv1D(8, 13, padding='valid', activation='relu', strides=1)(inputs)
conv1 = MaxPooling1D(3)(conv1)
conv1 = Dropout(0.2)(conv1)

# Second Conv1D Layer
conv2 = Conv1D(16, 11, padding='valid', activation='relu', strides=1)(conv1)
conv2 = MaxPooling1D(3)(conv2)
conv2 = Dropout(0.2)(conv2)

# Third Conv1D Layer
conv3 = Conv1D(32, 9, padding='valid', activation='relu', strides=1)(conv2)
conv3 = MaxPooling1D(3)(conv3)
conv3 = Dropout(0.3)(conv3)

# Fourth Conv1D Layer
conv4 = Conv1D(64, 7, padding='valid', activation='relu', strides=1)(conv3)
conv4 = MaxPooling1D(3)(conv4)
conv4 = Dropout(0.4)(conv4)

# Fifth Conv1D Layer
conv5 = Conv1D(128, 5, padding='valid', activation='relu', strides=1)(conv4)
conv5 = MaxPooling1D(3)(conv5)
conv5 = Dropout(0.5)(conv5)

# Flatten Layer
flatten = Flatten()(conv5)

# Dense Layer 1
dense1 = Dense(256, activation='relu')(flatten)
dense1 = Dropout(0.3)(dense1)

# Dense Layer 2
dense2 = Dense(128, activation='relu')(dense1)
dense2 = Dropout(0.4)(dense2)

# Dense Layer 3
dense3 = Dense(64, activation='relu')(dense2)
dense3 = Dropout(0.5)(dense3)

outputs = Dense(len(labels), activation='softmax')(dense3)

model = Model(inputs, outputs)
model.summary()
```

```python
#cross entropy
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
#Selecting the Best Model
es= EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=15, min_delta=0.0001)
```

```python
#Model fitting
history=model.fit(x_tr, y_tr, epochs=100,callbacks=es, batch_size=100, validation_data=(x_val,y_val))
```

# Testing


In a three-class classification problem where 85% of points are in class A, 10% in class B, and 5% in class C, achieving 85% accuracy means correctly identifying 85% of all data points across all classes. However, this high accuracy might be misleading, as it might primarily reflect the accurate identification of the majority class (Class A). For instance, a classifier could predict all samples as Class A and still attain 85% accuracy because of the imbalance. Multiclass classification results tend to be more intricate than binary classifications due to multiple classes, making it essential to consider other evaluation tools beyond accuracy. The confusion matrix and classification report are valuable tools in this scenario. The confusion matrix provides a detailed breakdown of correct and incorrect predictions for each class, shedding light on where the classifier excels or struggles. The classification report offers additional insights, presenting metrics like precision, recall, and F1-score for each class, offering a more nuanced understanding of the model's performance across all classes.

Precision serves as a metric that assesses the accuracy of the model's positive predictions. It is derived from the ratio between true positive predictions and the total of true positives and false positives. 

Recall measures the model's proficiency in recognizing all relevant instances. It's computed as the ratio of true positive predictions to the combined sum of true positives and false negatives. Essentially, recall underscores the model's effectiveness in capturing all actual positive instances, prioritizing their identification even at the cost of some misclassifications.

The multiclass f-score is a prevalent metric for imbalanced datasets in a multiclass setting. It computes individual binary f-scores for each class, treating each class as the positive one while considering the rest as negatives. Here are the common averaging strategies used with the multiclass f-score:

"Macro" averaging: This calculates the unweighted mean of the per-class f-scores. Each class holds equal weight regardless of its size.

"Weighted" averaging: This computes the mean of per-class f-scores, taking into account the support (number of instances) for each class. It's the metric typically reported in a classification report.

"Micro" averaging: This approach totals the false positives, false negatives, and true positives across all classes and then computes precision, recall, and f-score based on these counts.

Choosing the appropriate averaging strategy depends on the emphasis placed on the samples or classes: If equal emphasis is given to each sample, the "micro" average f1-score is recommended. If equal emphasis is given to each class, the "macro" average f1-score is recommended. (Müller and Guido, 2017)
